```python
import mysql.connector as con
import os
```

```python
db_client = con.Connect(host='localhost',user='root',passwd = 'mysqlpass',use_pure = True)
```

```python
cursor = db_client.cursor()
```

```python
cursor.execute("show databases")
cursor.fetchall()
```

```python
#2 create a database called as cardataset
table_name = 'car_data'
database_name = 'sqltask'
cursor.execute("create database {}".format(database_name))
cursor.execute("show databases")
cursor.fetchall()
```

```python
#3 create a table called as car with a column name given in dataset description
cursor.execute("create table sqltask.car_data(buying VARCHAR(30), maint VARCHAR(30),doors VARCHAR(30),persons VARCHAR(30),lug_boot VARCHAR(30), safety VARCHAR(30),class VARCHAR(20))")
cursor.execute("use sqltask")
cursor.execute("show tables")
cursor.fetchall()
```

```python
#4 Dump all the data into car table
import csv
with open('car_data.txt','r') as f:
    rows = csv.reader(f,delimiter='\n')
    for row in rows:
        data = row[0].split(',')
        query = "insert into sqltask.car_data values(%s,%s,%s,%s,%s,%s,%s)"
        cursor.execute(query,data)
db_client.commit()
        
```

```python
cursor.execute("select* from car_data")
cursor.fetchall()
```

```python
#5 try to check weather all the data is aviable inside your table or not
with open('car_data.txt','r') as f:
    rows = csv.reader(f,delimiter='\n')
    count = 0
    for row in rows:
        count +=1
print(count)
```

 Executes query ---> select count(*) from car_data;  in sql workbench gave result to 1728. Hence all the records are there in the database

```python
#6 try to group all the data with COL1 AND count occurences of each and every record based on col1 value 
cursor.execute("select buying,count(buying) from car_data group by buying")
cursor.fetchall()
```

```python
#7 Try to filter a record where col 3 value will be 4 .
cursor.execute("select * from car_data where doors=4 ")
cursor.fetchall()
```

```python
#8 Try to update a col 3 value with 8 whereever you have value equal to 2 
cursor.execute("update car_data set doors='8' where doors='2'")
db_client.commit()
cursor.execute("select * from car_data where doors='8'")
cursor.fetchall()
```

```python
#9 Drop table
cursor.execute("drop table sqltask.car_data")
db_client.commit()

```

```python
#delete database
cursor.execute("drop database sqltask")
db_client.commit()
```
