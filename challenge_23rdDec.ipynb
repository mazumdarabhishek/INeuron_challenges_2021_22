```python
#20 appending two PDF files together
import PyPDF2 as pypd
def pdf_merger(f_1='',f_2=''):
    file_1 = open(f_1,'rb')
    file_2 = open(f_2,'rb')
    reader_1 = pypd.PdfFileReader(file_1)
    reader_2 = pypd.PdfFileReader(file_2)
    writer = pypd.PdfFileWriter()
    for page in range(reader_1.numPages):
        p_obj = reader_1.getPage(page)
        writer.addPage(p_obj)

    for page in range(reader_2.numPages):
        p_obj = reader_2.getPage(page)
        writer.addPage(p_obj)
    merged_file = open("Mergerd_file.pdf",'wb')
    writer.write(merged_file)
    return 'Merger successful new merged file created'


```

```python
f_1='data science interview questions-1.pdf'
f_2 = 'Data science interview questions-2.pdf'
pdf_merger(f_1,f_2)
```
